명화를 읽어 주는 도슨트 만들기

```python
from dotenv import load_dotenv
load_dotenv()

from google import genai

client = genai.Client()
```

```python
user_input =("내 이름은 최원석이야")
```

```python
response = client.models.generate_content(
    model="gemini-3-flash-preview",
    contents=user_input,
)

print(response.text)
```
