```python
from pydantic import BaseModel, Field

#스키마
class Movie(BaseModel):
    """영화에 대한 세부 정보"""
    title: str = Field(..., description="영화의 제목")
    year: int = Field(..., description="영화의 개봉된 연도")
    director: str = Field(..., description="영화의 감독")
    reting: float = Field(description="10점 만점 영화 평점")
```

```python
model_with_structure = model.with_structured_output(Movie)
```

```python
res = model_with_structure.invoke("영화 기생충에 대한 세부 정보를 제공해줘")
```
